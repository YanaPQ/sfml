```python
# пропуски данных есть в столбцах, перечисленных ниже:

# строки с отсутствующиеи данными в следующих столбцах удалим:

# ORG_TP_STATE                  27420 non-null object форма собственности компании
# ORG_TP_FCAPITAL               27425 non-null object отношение к иностранному капиталу
# JOB_DIR                       27420 non-null object направление деятельности внутри компании
# TP_PROVINCE                   29543 non-null object область торговой точки, где клиент брал послед...
# GEN_INDUSTRY                  27420 non-null object отрасль работы клиента
# GEN_TITLE                     27420 non-null object должность клиента

# пропущенные значения заполним медианными:
# WORK_TIME                     27416 non-null float64 время работы на текущем месте (в месяцах)

# следующие данные удалим как столбец, т.к. при удалении их как строк потеряем слишком много данных:
# PREVIOUS_CARD_NUM_UTILIZED    600 non-null float64 количество уже утилизированных карт ( если пус...
```

```python
df = df[~df['ORG_TP_STATE'].isnull()]
df = df[~df['ORG_TP_FCAPITAL'].isnull()]
df = df[~df['JOB_DIR'].isnull()]
df = df[~df['TP_PROVINCE'].isnull()]
df = df[~df['GEN_INDUSTRY'].isnull()]
df = df[~df['GEN_TITLE'].isnull()]
```
